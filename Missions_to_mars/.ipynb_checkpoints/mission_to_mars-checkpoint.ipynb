```python
from bs4 import BeautifulSoup
import requests
import os
from splinter import Browser
import pandas as pd
import pymongo
from pprint import pprint
```

```python
#news
```

```python
url = 'https://mars.nasa.gov/news/'
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
news_title = soup.find('div', class_ = 'content_title').text
print(news_title)
```

```python
news_content = soup.find('div', class_ = 'article_teaser_body').text
print(news_content)
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(url)
# browser.click_link_by_partial_text('FULL ')
html = browser.html
soup = BeautifulSoup(html, 'html.parser')

```

```python
fancybox = soup.find_all('a', class_='fancybox')[0]
fea_img = fancybox.get('data-fancybox-href')
```

```python
url = 'https://twitter.com/marswxreport?lang=en'
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
weather = soup.find('p', class_ = 'TweetTextSize TweetTextSize--normal js-tweet-text tweet-text').text
```

```python
weather = weather[0:-26]
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.marsDB
```

```python
collection = db.data
```

```python
docs = {
        "n_t" : news_title,
        "n_c" : news_content,
        "f_i": fea_img,
        "wea" : weather,
        "ta" : table,
        "h_i_1": hem_img_1,
        "h_i_2": hem_img_2,
        "h_i_3": hem_img_3,
        "h_i_4": hem_img_4}
```

```python
collection.insert_many(docs)
```
